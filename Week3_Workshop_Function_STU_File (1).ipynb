Create a function that calculates the area of a circle and returns it.

`(Area = 3.14 √ó radius √ó radius)`

**Output:**
```
Circle area: 78.5
Larger circle area: 314.0
Combined area: 78.26

````python
# Solution:
def circle_area(radius):

    area = 3.14 * radius * radius
    return area


area1 = circle_area(10)
area2 = circle_area(5)
area3 = circle_area(10.0)

print(f"Circle area: {area1}")
print(f"Larger circle area: {area2}")
print(f"Combined area: {area3}")
````

**Task: Create a function that calculates total and average from three game scores and returns both.**

```

Output:

Total: 270, , Average: 90.0
Your total is 280 and average is 93.33333333333333
```

````python
# Solution:

def win_reslut(score1, score2, score3):

    total = score1 + score2 + score3
    average = total / 3
    return total, average

total1, average1 = win_reslut(80, 80, 80)
print(f"Total: {total1}, Average: {average1}")


total2, average2 = win_reslut(90, 90, 90)
print(f"Your total is {total2} and average is {average2}")
````
