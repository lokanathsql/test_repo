**Task: Write a program in Python that asks the user to enter an integer n and display the value of the sum 1 + 2 + ... + n**

```python
#Solution:

n = int(input("Enter a integer: "))

sum = 0
for i in range(1, n + 1):
    sum += i

print(f"The sum 1+2 ..+{n}: {sum}")
```

**Task: Python Program for Sum of squares of first n natural numbers [use function]**

```python
#Solution:

def Sum_of_Squares(n):
    total = 0
    for i in range(1, n + 1):
        total = total + (i * i)
    return total

n = int(input("Give a number: "))

result = Sum_of_Squares(n)

print("Sum of squares is:", result)

```

**Task: Write a program to calculate the sum of numbers from 1 to 5, showing the intermediate and final totals.**

```python
#Solution:
def Sum_of_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
        print(f"intermediate total:{total}")
    return total

result_sum = Sum_of_numbers(5)
print(f"final total: {result_sum}")
```
