**Palindrome Checker**

**Task: Write a function is_palindrome(text) that returns True if the string is the same when reversed, otherwise False.**

````python
# Solution:

def is_palindrome(text):
    return text == text[::-1]


print(is_palindrome("madam"))
print(is_palindrome("racecar"))
print(is_palindrome("hello"))

#text[::-1] → reverses the string
#text == text[::-1] → checks if the original string is equal to its reversed version
#It returns True if they are the same (palindrome), otherwise False
````

**Multiple Return Values**

**Task: Create a function circle_measurements(radius) that returns both area and circumference of a circle.**

**Use formulas:**

**Area = π × r²**

**Circumference = 2 × π × r**

````python
# Solution:

import math

def circle_measurements(radius):
    area = math.pi * radius ** 2
    circumference = 2 * math.pi * radius
    return area, circumference

# Example
a, c = circle_measurements(5)

print("Area:", a)
print("Circumference:", c)
````

**Task: Create a function that calculates total and average from three game scores and returns both.**

```

Output:

Total: 270, , Average: 90.0
Your total is 280 and average is 93.33333333333333
```

````python
# Solution:

def win_reslut(score1, score2, score3):

    total = score1 + score2 + score3
    average = total / 3
    return total, average

total1, average1 = win_reslut(80, 80, 80)
print(f"Total: {total1}, Average: {average1}")


total2, average2 = win_reslut(90, 90, 90)
print(f"Your total is {total2} and average is {average2}")
````

**Task: Create a function that finds both the smallest and largest of three numbers.**
```
Output:

Smallest: 5, Largest: 12
Range: 25 to 100
```

````python
# Solution:

def find_smallest_largest(a, b, c):
    smallest = min(a, b, c)
    largest = max(a, b, c)
    return smallest, largest

# Example 1
small, large = find_smallest_largest(5, 12, 9)
print(f"Smallest: {small}, Largest: {large}")

# Example 2
small, large = find_smallest_largest(25, 100, 50)
print(f"Range: {small} to {large}")
````

**Task: Create a function that calculates the average of any amount of numbers**

````python
# Solution:
def calculate_average(*numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)


print("Average of 10, 20, 30:", calculate_average(10, 20, 30))
print("Average of 5, 15:", calculate_average(5, 15))
````
